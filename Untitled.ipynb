```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx, yy = np.arange(-0.6, 0.6, 0.01), np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(xx, yy)
g = X ** 2 + ((Y**4) - 0.5 * (Y**2))
```

```python
fig, ax = plt.subplots(1)
ax.set_aspect('equal')
ax.contourf(X, Y, np.clip(g, a_max=0.25, a_min=None), levels=5, cmap='Greys_r')
ax.contour(X, Y, np.clip(g, a_max=0.25, a_min=None), levels=5, cmap='Greys')
fig.savefig('foo.png', dpi=300)
```
